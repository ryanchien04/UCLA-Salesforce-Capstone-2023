```python
import ast, string, random

N = 7
dictionary_holder = {}

# obfuscate
def generate_random_string(receivedIdentifier):
    if(receivedIdentifier == "print"):
      return receivedIdentifier
    elif receivedIdentifier in dictionary_holder.keys():
      return dictionary_holder[receivedIdentifier]
    else:
      new_arg_name = ''.join(random.choices(string.ascii_uppercase + string.digits, k=N))
      dictionary_holder.update({receivedIdentifier: new_arg_name})
      return new_arg_name

# lower
def generate_lower_string(receivedIdentifier):
    if(receivedIdentifier == "print"):
      return receivedIdentifier
    elif receivedIdentifier in dictionary_holder.keys():
      return dictionary_holder[receivedIdentifier]
    else:
      new_arg_name = receivedIdentifier.lower()
      dictionary_holder.update({receivedIdentifier: new_arg_name})
      return new_arg_name

# upper
def generate_upper_string(receivedIdentifier):
    if(receivedIdentifier == "print"):
      return receivedIdentifier
    elif receivedIdentifier in dictionary_holder.keys():
      return dictionary_holder[receivedIdentifier]
    else:
      new_arg_name = receivedIdentifier.upper()
      dictionary_holder.update({receivedIdentifier: new_arg_name})
      return new_arg_name

# character subset - 2
def generate_char_2_substring_string(receivedIdentifier):
    if(receivedIdentifier == "print"):
      return receivedIdentifier
    elif receivedIdentifier in dictionary_holder.keys():
      return dictionary_holder[receivedIdentifier]
    else:
      new_arg_name = receivedIdentifier[0:2] if len(receivedIdentifier) > 2 else receivedIdentifier
      dictionary_holder.update({receivedIdentifier: new_arg_name})
      return new_arg_name

# character subset - 3
def generate_char_3_substring_string(receivedIdentifier):
    if(receivedIdentifier == "print"):
      return receivedIdentifier
    elif receivedIdentifier in dictionary_holder.keys():
      return dictionary_holder[receivedIdentifier]
    else:
      new_arg_name = receivedIdentifier[0:3] if len(receivedIdentifier) > 3 else receivedIdentifier
      dictionary_holder.update({receivedIdentifier: new_arg_name})
      return new_arg_name

class transformerClass(ast.NodeTransformer):
    def visit_arg(self, node):
        # node.arg.lower()
        return ast.arg(**{**node.__dict__, 'arg':generate_char_3_substring_string(node.arg)})

    def visit_Name(self, node):
        # node.id.lower()
        return ast.Name(**{**node.__dict__, 'id':generate_char_3_substring_string(node.id)})

code = """
def simple_interest(principal,time_period,rate_of_interest):
    print('The principal is', principal)
    print('The time period is', time_period)
    print('The rate of interest is',rate_of_interest)

    simple_interest_calc_result = (principal * time_period * rate_of_interest)/100

    print('The Simple Interest is', simple_interest_calc_result)
    return simple_interest_calc_result
"""
print(ast.unparse(transformerClass().visit(ast.parse(code))))
```
